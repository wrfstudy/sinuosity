```python
fig, ax1 = plt.subplots()
width = 0.3
bars1 = ax1.bar(x[0:2], [43.4,25.4], width=width, color=col1, label='Period I')
bars2 = ax1.bar(x[0:2] + width, [60.8, 39.2], width=width, color=col2, label='Period II')
for bar in bars1:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.1f}%', ha='center', va='bottom', fontsize=12)
for bar in bars2:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.1f}%', ha='center', va='bottom', fontsize=12)
ax1.set_xticks([x[0] + width / 2, x[1] + width / 2, x[2] + width / 2, x[3] + width / 2])
ax1.set_xticklabels(ax1.get_xticklabels())  # Change of x title
ax1.set_ylabel('Percentage', fontsize=15)
ax1.set_ylim(0,70)
ax1.legend(loc='upper right', bbox_to_anchor=(0.75, 1),fontsize=10)

ax2 = ax1.twinx()
bars3 = ax2.bar(x[2:4], [4.41, 2.59], width=width, color=col3, label='Period I')  # 黄色
bars4 = ax2.bar(x[2:4] + width, [5.83, 3.76], width=width, color=col4, label='Period II')  # 蓝色
for bar in bars3:
    yval = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.2f}', ha='center', va='bottom', fontsize=12)
for bar in bars4:
    yval = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.2f}', ha='center', va='bottom', fontsize=12)
ax2.legend(fontsize=10,loc='upper left', bbox_to_anchor=(0.74, 1))
ax2.set_ylabel('Annual Mean Frequency', fontsize=15)
ax2.set_ylim(0, 10)
plt.title('(c) Turing-Type TCs in box', fontsize=15, loc='left')
names=['Total TCs','Landfalling TCs',
       'Total TCs', 'Landfalling TCs']
plt.xticks(x + width/2,labels=names)
plt.show()
```
