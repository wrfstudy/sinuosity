```python
df1 = pd.read_excel(f'Typhoon_Cluster_3.xlsx')
```

```python
indallc1=np.array(df1['Typhoon ID'])
```

```python
indallc1.shape
```

```python
timec1=isotime[indallc1]
timec1_datetime = np.array([np.datetime64(time.decode('utf-8')) for time in timec1.flatten() if time])
```

```python
allcape=[]
for i in range(1982,2024):
    print(i)
    data= xr.open_dataset(f'E:/data/daily2.5/cape/cape_{i}.nc')
    recape =data['cape']*1#这样后面计算会快一点
    res_time = data['valid_time']#需要的时间值
    common_times = np.intersect1d(timec1_datetime, res_time.values)
    capevalue=recape.sel(valid_time=common_times)
    allcape.append(capevalue)
```

```python
# 将存储在vwss中的1981-1983年的vws数据进行平均
cape_1 = allcape[:17]  # 提取1981到1983年的vws数据
cape_2 = allcape[17:] 
```

```python
data
```

```python
# 将三年的数据沿时间维度进行叠加，然后求平均
cape_p1 = xr.concat(cape_1, dim='valid_time')
cape_p2 = xr.concat(cape_2, dim='valid_time')
```

```python
mecape_1 = cape_p1.mean(dim='valid_time')
mecape_2 = cape_p2.mean(dim='valid_time')
```

```python
cape_diff = mecape_2 - mecape_1
```

```python
latcape = data['latitude']
loncape = data['longitude']
```

```python
from scipy.stats import ttest_ind

p_values = np.zeros((len(latcape), len(loncape)))

for i in range(len(latcape)):
    for j in range(len(loncape)):
        t_stat, p_values[i, j] = ttest_ind(cape_p1[:, i, j], cape_p2[:, i, j], equal_var=False)

# 显著性水平低于0.05的地方标记为True
significant_cape = p_values < 0.05# 假设 tas_period1 和 tas_period2 是两个时间段的数据
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree(central_longitude=180))
ax.set_extent([100,240, 0,60], crs=ccrs.PlateCarree())

# 绘制网格线
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')

vmin = -300
vmax = 300

# Create contour plot
contour = ax.contourf(loncape, latcape, cape_diff, cmap='coolwarm', levels=np.linspace(vmin, vmax, 20), extend='both',transform=ccrs.PlateCarree())

# 在显著性通过95%的地方打黑点
significant_points_cape = np.where(significant_cape)
ax.scatter(loncape[significant_points_cape[1]], latcape[significant_points_cape[0]], color='black', s=1, transform=ccrs.PlateCarree())

cbar = plt.colorbar(contour, ax=ax, orientation='vertical', label='vws')

plt.show()
```
